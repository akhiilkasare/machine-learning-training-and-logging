### Data Analysis

```python
print("The length of train dataset : ",train.shape)
print("The length of test dataset : ",test.shape)
```

```python
## Checking for categorical columns

categorical_col = []
for column in train.columns:
    if train[column].dtype == object and len(train[column].unique()) <= 50:
        categorical_col.append(column)
        print(f"{column} : {train[column].unique()}")
        print("====================================")
```

```python
## Checking for numerical columns

numerical_col = []
for column in train.columns:
    if train[column].dtype != object and len(train[column].unique()) <= 50:
        numerical_col.append(column)
        print(f"{column} : {train[column].unique()}")
        print("====================================")
```

```python
sns.countplot(train['Response'])
```

```python
### Replacing the + with space
```

```python
def replace_percent(x):
    try: 
        return x.replace('+', '')
    except AttributeError:
        return np.NaN

train.Holding_Policy_Duration = train.Holding_Policy_Duration.map(replace_percent)
```

```python
### Checking for null values
```

```python
percent_missing = train.isnull().sum() * 100 / len(train)
# print(percent_missing)
missing_values = pd.DataFrame({'column_name' : train.columns,
                              '% missing values' : percent_missing})
missing_values
```

```python
cols = ['City_Code', 'Accomodation_Type', 'Reco_Insurance_Type']

train = pd.get_dummies(train, columns=cols)

```

```python
train.head()
```

```python
train['Holding_Policy_Duration'] = train['Holding_Policy_Duration'].astype('float64')
```
